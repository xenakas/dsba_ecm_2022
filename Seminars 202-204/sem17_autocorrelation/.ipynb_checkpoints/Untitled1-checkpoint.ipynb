```r
c1 <- ts(data['COSM'], frequency = 4, start=1990)
d1 <- ts(data['DPI'], frequency = 4, start=1990)
```

```r
library('dynlm')
```

```r
mod3 <- dynlm(c1~d1+L(c1))
```

```r
summary(mod3)
```

```r
dwtest(mod3)
```

```r
test <- dwtest(mod3)
sum <- summary(mod3)$coefficients
sum
```
