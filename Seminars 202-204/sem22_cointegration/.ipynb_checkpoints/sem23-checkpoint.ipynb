```r
library(hexView)
```

```r
data <- readEViews('EXPEND.WF1')
```

```r
library(forecast)
```

```r
x <- ts(data['GASO'], start=1990)
ggtsdisplay(x)
```

```r
library(urca)
```

```r
summary(ur.df(x, type='trend'))
```

```r
ggtsdisplay(diff(x))
```

```r
summary(ur.df(diff(x), type='trend'))
```

```r
ggtsdisplay(diff(diff(x)))
```

```r
summary(ur.df(diff(diff(x)), type='trend'))
```

```r
x_c <- ts(data['COSM'], start=1990)
x_d <- ts(data['DPI'], start=1990)
x_p <- ts(data['PRELCOSM'], start=1990)
```

```r
ggtsdisplay(x_c)
ggtsdisplay(x_d)
ggtsdisplay(x_p)
```

```r
summary(ur.df(x_c, type='drift'))
summary(ur.df(x_d, type='drift'))
summary(ur.df(x_p, type='drift'))
```

```r
summary(ur.df(diff(x_c), type='drift'))
summary(ur.df(diff(x_d), type='drift'))
summary(ur.df(diff(x_p), type='drift'))
```

```r
mod1 <- lm(x_c~x_d+x_p)

summary(mod1)
```

```r
ggtsdisplay(mod1$residuals) 
```

```r
summary(ur.df(mod1$residuals, type='none'))
```

```r
mod1 <- lm(diff(x_c)~diff(x_d)+diff(x_p))

summary(mod1)
```

```r
plot(x_c)
lines(ts(x_c[1]+diffinv(mod1$fitted.values),1990), type='l', col='red')
```
