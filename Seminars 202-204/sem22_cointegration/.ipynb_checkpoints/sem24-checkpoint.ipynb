```r
library(hexView)
```

```r
data <-readEViews('EXPEND.WF1')
```

```r
cosm <- ts(data['COSM'], frequency = 1, start = 1990)
dpi <- ts(data['DPI'], frequency = 1, start = 1990)
pr <- ts(data['PRELCOSM'], frequency = 1, start = 1990)
```

```r
mod2 <- lm(diff(cosm)~0+diff(dpi)+diff(pr))
summary(mod2)
```

```r
plot(mod2$fitted)
```

```r
plot(seq(1,25), data[['COSM']],type='l')
lines(cosm[1]+diffinv(mod2$fitted), col='red')
```

```r
#ECM Step 1 
mod1 <- lm(cosm~dpi+pr)
res <- mod1$residuals
#ECM Step 2
library('dynlm')
mod3 <- dynlm(diff(cosm)~L(res)[2:25]+diff(dpi)+diff(pr))
```

```r
summary(mod3)
```

```r
mod3 <- lm(diff(cosm)~c(NA,res[1:23])+diff(dpi)+diff(pr))
summary(mod3)
```

```r
plot(mod3$fitted.values)
```

```r
# mod3$fitted.values
```

```r
# plot(seq(1,25), data[['COSM']],type='l')
plot(cosm,type='l')
lines(ts(cosm[1]+diffinv(mod2$fitted), start=1990), col='red')
lines(cosm[1]+diffinv(mod3$fitted.values), col='green')
```

```r
library('lmtest')
```

```r
grangertest(data[['PRELDOC']], data[['DOC']], order=2)
#H0: PRELDOC doesn't Grnager cause DOC

grangertest(data[['DOC']], data[['PRELDOC']], order=2)
#H0: DOC doesn't Grnager cause  PRELDOC 
```

```r
df1 <- data[,c('COSM', 'DPI', "PRELCOSM")]
```

```r
library(vars)
```

```r
VARselect(df1, lag.max = 5, type='const')
```

```r
summary(VAR(df1, p=5))
```

```r
summary(VAR(df1, p=2))
```

```r
mod1 <- VAR(df1, p=2)
forecast<-predict(mod1, n.ahead = 12, ci=0.95)
```

```r
fanchart(forecast, names='COSM')
fanchart(forecast, names='DPI')
fanchart(forecast, names='PRELCOSM')
```
