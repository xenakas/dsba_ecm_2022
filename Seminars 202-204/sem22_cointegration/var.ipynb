```r
cosm<-ts(data['COSM'], frequency = 1, start=1990)
```

```r
dpi<-ts(data['DPI'], frequency = 1, start=1990)
pr<-ts(data['PRELCOSM'], frequency = 1, start=1990)
```

```r
summary(ur.df(cosm, type='trend', selectlags='AIC'))
summary(ur.df(dpi, type='trend', selectlags='AIC'))
summary(ur.df(pr, type='trend', selectlags='AIC'))
```

```r
ggtsdisplay(cosm)
ggtsdisplay(dpi)
ggtsdisplay(pr)
```

```r
ggtsdisplay(diff(cosm))
ggtsdisplay(diff(dpi))
ggtsdisplay(diff(pr))
```

```r
summary(ur.df(diff(cosm), type='trend', selectlags='AIC'))
summary(ur.df(diff(dpi), type='trend', selectlags='AIC'))
summary(ur.df(diff(pr), type='trend', selectlags='AIC'))
```

```r
mod1 <- lm(cosm~dpi+pr)
res1 <- mod1$residuals
```

```r
summary(ur.df(res1, type='none', selectlags='AIC'))
```

```r
summary(lm(cosm~dpi+pr))
```

```r
summary(lm(diff(cosm)~diff(dpi)+diff(pr)))
```

```r
mod2<-lm(diff(cosm)~0+diff(dpi)+diff(pr))
summary(mod2)
```

```r
# data[['COSM']]
plot(seq(1,25),data[['COSM']], type='l')
lines(cosm[1]+diffinv(mod2$fitted), type='line', col='red')
```

```r
mod1 <- lm(cosm~dpi+pr)
res <- mod1$residuals
```

```r
library(dynlm)
```

```r
mod3<-dynlm(diff(cosm)~L(res)[2:25]+diff(dpi)+diff(pr))
summary(mod3)
```

```r
mod3$fitted.values
```

```r
# data[['COSM']]
plot(seq(1,25),data[['COSM']], type='l')
lines(cosm[1]+diffinv(mod2$fitted), type='line', col='red')
# lines(cosm[1]+diffinv(mod3$fitted.values), type='line', col='green')
```

```r
plot(cosm)
lines(cosm[1]+diffinv(mod3$fitted.values), type='line', col='green')
lines(ts(cosm[1]+diffinv(mod2$fitted), start=1990), type='line', col='red')
```
